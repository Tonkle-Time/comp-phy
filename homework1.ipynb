```python
# Example 1
#I googled how to get started and other vague code lines like plt. functions
import numpy as np
#Import the numpy library 
import matplotlib.pyplot as plt
#Import matplotlib library
# Generate 100 equally spaced data points in the interval [0, 2]
xn = np.linspace(0, 2 * np.pi, 100)
# Evaluate f(xn) = exp[6 * sin(xn)]
fn = np.exp(6 * np.sin(xn))
#These two lines of code were suggested by AI

#Plot data represented by red diamonds that are disconnected
plt.figure(figsize=(10,5))
plt.title('Problem 1 A')
plt.plot(xn, fn, 'rD', linestyle='None')
#this took some effort to figure out it is the x and y values to be plotted
#'rD' is a format string that specifies the marker style and color
# and linestyle='None' tells Matplotlib not to connect the points with lines
plt.xlabel('x')
plt.ylabel('f(x) = exp(6 sin(x))')
plt.grid(True)
plt.tight_layout() # was recommended to automatically 
#adjusts the spacing between plot elements
plt.show()

#Plot data continuous green curve
plt.figure(figsize=(10, 5))
plt.plot(xn, fn, 'g-')  # 'g-' means green solid line
plt.xlabel('x')
plt.ylabel('f(x) = exp(6 sin(x))')
plt.title('Problem 1 B')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Problem 2
import numpy as np
import matplotlib.pyplot as plt

#Load the data from the file
data = np.loadtxt('NoisyFunction.dat')

# Split the data into t and y arrays - had to google this as well
tn = data[:, 0]       # First column: time values
yn = data[:, 1]       # Second column: y(tn) values

#Plot 1
plt.figure(figsize=(10, 5))
plt.scatter(tn, yn, color='blue', marker='^')  
plt.xlabel(r'$t_n$') #googled how to do subscripts
plt.ylabel(r'$y(t_n)$')
plt.title('Problem 2 A')
plt.grid(True)
plt.tight_layout()
plt.show()

#AI recommended this although the cell ran without it?
epsilon = 1e-10
log_y2 = np.log(yn**2 + epsilon)
 

#Plot 2
plt.figure(figsize=(10, 5))
plt.scatter(tn, log_y2, color='purple', marker='s')  
plt.xlabel(r'$t_n$')
plt.ylabel(r'$\log(y(t_n)^2)$')
plt.title('Problem 2 B')
plt.grid(True)
plt.tight_layout()
plt.show()
```
